```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer
#ps = PorterStemmer()
```

```python
df=pd.read_csv('sms collection.csv',sep='\t',names=['Label','Message'])
```

```python
df
```

# Data Cleaning and Preprocessing

```python
df.isna().sum()
```

```python
df['Message']=df['Message'].apply(lambda x : x.lower())
```

```python
corpus=[]
for i in range(0,len(df)):
    review=re.sub('[^a-zA-z]',' ',df['Message'][i])
    review=review.lower()
    review=review.split()
    review=[word for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

# Create bag of words

```python
cv=CountVectorizer(binary=True,max_features=100)
```

```python
x=cv.fit_transform(corpus).toarray()
```

```python
x[1]
```

```python
x.shape
```

# Create a bow using lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(df)):
    review=re.sub('[^a-zA-z]',' ',df['Message'][i])
    review=review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word,pos='r') for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
cv_2=CountVectorizer(max_features=200,binary=True)
```

```python
x=cv_2.fit_transform(corpus).toarray()
```

```python
x
```

# N-Grams

```python
cv_ngram=CountVectorizer(max_features=2500,ngram_range=(1,3),binary=True)
x=cv_ngram.fit_transform(corpus).toarray()
x
```
